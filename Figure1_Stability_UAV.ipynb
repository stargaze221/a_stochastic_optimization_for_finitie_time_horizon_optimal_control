```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import pickle

fig, ax = plt.subplots()
    



### Add the rect. obs. ###
LB_X = 4
UB_X = 6
LB_Y = 4
UB_Y = 8

ax.add_patch(Rectangle((LB_X, LB_Y), UB_X - LB_X, UB_Y - LB_Y))

### Load files and add trajectories ###
# list_f_name = ['UAV_MPPI_-0.5', 'UAV_MPPI_-0.1', 'UAV_MPPI_0', 'UAV_MPPI_0.1', 'UAV_MPPI_0.5'] #, 'UAV_MPPI_0.5']

# for f_name in list_f_name:
#     x,y = pickle.load( open( f_name+".p", "rb" ) )
#     ax.plot(x,y, label="f_name")

list_str_a_s = ['-0.5', '-0.1', '0', '0.1', '0.5']
for a_str in list_str_a_s:
    x,y = pickle.load( open( 'UAV_MPPI_'+a_str+".p", "rb" ) )
    ax.plot(x,y, label='a='+a_str)
    
### Add goal positions ###
ax.plot(8, 8, 'b+')
    
    
ax.grid()
ax.legend()
ax.set_aspect('equal')
ax.set_xlim(-2, 10)
ax.set_ylim(-2, 10)
plt.legend(loc='lower right')
plt.tight_layout()
fig.savefig('figures/figure1.pdf', format='pdf')
fig.savefig('figures/figure1.png')
plt.show()
```
